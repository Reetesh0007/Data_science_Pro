```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Load the dataset
titanic_data = pd.read_csv('titanic_dataset.csv')

# Display the first few rows of the dataset
print(titanic_data.head())
```

```python
plt.hist(titanic_data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Count by Sex')
plt.legend(title='Sex', loc='upper right')
plt.show()
```

```python
sns.catplot(data=titanic_data, x="Age", y='Sex', order=["Male", "Female"])
```

```python
sns.boxplot(x='Pclass', y='Fare', data=titanic_data)
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.title('Fare Distribution by Passenger Class')
plt.show()
```

```python
sns.catplot(data=titanic_data, x="Pclass", y="Fare", hue="Sex", kind='box')
plt.show()
```

```python
#visualization using violin
sns.catplot(data=titanic_data, x="Pclass", y="Fare", hue="Sex", kind='violin')
plt.show()
```

```python
#visualization using violin
sns.catplot(data=titanic_data, x="Pclass", y="Fare", hue="Sex", kind='violin',split=True)
plt.show()
```

```python
#visualization using violin
sns.catplot(data=titanic_data, x="Pclass", y="Fare", hue="Sex", kind='violin',inner='quart', palette='pastel', split=True)
plt.show()
```

```python
sns.countplot(x='Embarked', data=titanic_data, palette='pastel')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Count of Passengers by Embarked Port')
plt.show()
```

```python
#Visualization using Heatmap
correlation_matrix = titanic_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data = pd.read_csv('titanic_dataset.csv', encoding='latin1')
print(titanic_data.info())
```

```python
print(titanic_data.describe())
```

```python
print(titanic_data.isnull().sum())
```

```python
# Distribution of passengers by gender
#visuals using Bar chart
sns.countplot(x='Sex', data=titanic_data)
plt.title('Distribution of Passengers by Gender')
plt.show()
```

```python
# Distribution of passengers by passenger class
#visuals using Bar chart
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Distribution of Passengers by Class')
plt.show()
```

```python
# Distribution of passengers by survival
#visuals using Bar chart
sns.countplot(x='Survived', data=titanic_data)
plt.title('Distribution of Passengers by Survival')
plt.show()
```

```python
# Distribution of age
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.show()
```

```python
# Distribution of fares
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data['Fare'], bins=30, kde=True)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.show()
```

```python
# Survival by passenger class
#visuals using Bar chart
sns.countplot(x='Survived', hue='Pclass', data=titanic_data)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
sns.catplot(data=titanic_data.sort_values("Survived"), x='Survived', y='Pclass', kind='boxen')
plt.show()
```

```python
#visualization using violin
sns.catplot(data=titanic_data.sort_values("Survived"), x='Survived', y='Pclass', kind='violin')
plt.show()
```

```python
# Survival by gender
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
plt.title('Survival by Gender')
plt.show()
```

```python
# Survival by number of siblings/spouses aboard
sns.countplot(x='Survived', hue='SibSp', data=titanic_data)
plt.title('Survival by Siblings/Spouses Aboard')
plt.show()
```

```python
# Survival by number of parents/children aboard
sns.countplot(x='Survived', hue='Parch', data=titanic_data)
plt.title('Survival by Parents/Children Aboard')
plt.show()
```

```python
# Survival by embarked port
sns.countplot(x='Survived', hue='Embarked', data=titanic_data)
plt.title('Survival by Embarked Port')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(titanic_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
