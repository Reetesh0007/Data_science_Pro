# Import the library packages and  Visualize


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

```

# Checking the row head of data

```python

data = pd.read_csv('Sales_prediction.csv')
print(data.head())
```

# Checking the data null

```python

print(data.isnull().sum())
```

# Checking the correlation between features before start  training a machine learning model to predict future sales using heat map.

```python


plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr())
plt.show()
```

# Prepare the data to fit into a machine learning model then use a linear regression algorithms to train a sales prediction model.


```python

x = np.array(data.drop(["Sales"], 1))
y = np.array(data["Sales"])
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)

data = pd.DataFrame(data={"Predicted Sales": ypred.flatten()})
print(data)
```
